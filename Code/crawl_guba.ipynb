```python
# 导入所需的包

import requests

from bs4 import BeautifulSoup

import pandas as pd
```

```python
# 定义保存网页函数

def save(num):
    for i in range(1,num+1):
        url = f"http://guba.eastmoney.com/list,601099_{i}.html"

        res = requests.get(url)
        
        if res.status_code==200:
            res.encoding ="utf-8"
            with open(f"./data/guba/guba{i}.html","w",encoding="utf-8") as f:
                f.write(res.text)
```

```python
# 定义爬取函数

import re

import os
def crawl():

    k = 0
    df = pd.DataFrame(columns=["title","time"])
    dir_ls = os.listdir("./data/guba")
    for i in dir_ls:
        with open(f"./data/guba/{i}","r",encoding="utf-8") as f:
            html_text = f.read()
        
        # 生成BeautifulSoup对象
        soup = BeautifulSoup(html_text,"html.parser")
        div_label = soup.find("div",attrs={"id":"articlelistnew","class":"all hs_list"})
        # print(div_label)

        pattern = re.compile("articleh normal_post\s*\S*")

        div_ls = div_label.find_all("div",attrs={"class":pattern})

        for i in div_ls:
            span1 = i.find("span",attrs = {"class":"l3 a3"})
            title = span1.a.text
            span2 = i.find("span",attrs = {"class":"l5 a5"})
            time = span2.text
            df.loc[k] = [title,time]
            k+=1
    df.to_excel("./data/guba/guba.xlsx")
                        

        
```

```python
# 调用

# save(20)

crawl()
```
